```python
# DataFrame with list values
import pandas as pd

i = (10,20,30,40,50,60)
df = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with string values
import pandas as pd

i = ['string 1','string 2','string 3']
df = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with Dictionary values
import pandas as pd

i = {
    'Name':['Amol','Chetan','Nitesh','Laxmikant'],
    'Age':[23,25,24,25],
    'City':['Pune','Nashik','Mumbai','Latur']
}
df = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with Dictionary values
import pandas as pd

i = {
    'Sr.No.':[1,2,3,4],
    'Name':['Amol','Chetan','Nitesh','Laxmikant'],
    'Age':[23,25,24,25],
    'City':['Pune','Nashik','Mumbai','Latur']
}
df = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with Series
import pandas as pd

i = {'no':pd.Series([1,2,3,4,5],index=['a','b','c','d','e'])}
df  = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with Series 2 column
import pandas as pd

i = {'no1':pd.Series([1,2,3,4,5],index=['a','b','c','d','e']),
    'no2':pd.Series([6,7,8,9,0],index=['f','g','h','i','j'])}
df  = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with Series 2 column
import pandas as pd

i = {'no1':pd.Series([1,2,3,4,5],index=['a','b','c','d','e']),
    'no2':pd.Series([6,7,8,9,0],index=['a','b','c','d','e'])}
df  = pd.DataFrame(i)
print(df)
```

```python
# DataFrame with Series 2 column
import pandas as pd

i = {'no1':pd.Series([1,2,3,4,5],index=['a','b','c','d','e']),
    'no2':pd.Series([1,2,3,4,5],index=['f','g','h','i','j'])}
df  = pd.DataFrame(i)
print(df)
```

```python
# append() Method in DataFrame

i = pd.DataFrame({
    'X':[10,20,30,40],
    'Y':[5,15,25,35]
})

j = pd.DataFrame({
    'X':[50,60,70,80],
    'Y':[45,55,65,75]
})

i.append(j)
```

```python
# ignore_index in DataFrame

i = pd.DataFrame({
    'X':[10,20,30,40],
    'Y':[5,15,25,35]
})

j = pd.DataFrame({
    'X':[50,60,70,80],
    'Y':[45,55,65,75]
})

i.append(j, ignore_index=True)
```

```python
# ignore_index in DataFrame

i = pd.DataFrame({
    'X':[10,20,30,40],
    'Y':[5,15,25,35]
})

j = pd.DataFrame({
    'X':[50,60,70,80],
    'Y':[45,55,65,75]
})

i.append(j, ignore_index=False)
```

```python
# count() Method in DataFrame
import pandas as pd
import numpy 

i = pd.DataFrame({
    'Person':['John','Sam','Mac','Max'],
    'Age':[30,37,numpy.nan,35]
})

i.count()
```

```python
# column axis in DataFrame
import pandas as pd
import numpy 

i = pd.DataFrame({
    'Person':['John','Sam','Mac','Max'],
    'Age':[30,37,numpy.nan,35]
})

i.count(axis='columns')
```

```python
# column axis in DataFrame
import pandas as pd
import numpy 

i = pd.DataFrame({
    'Person':['John','Sam','Mac','Max'],
    'Age':[30,37,numpy.nan,35],
    'City':[numpy.nan,'Pune',numpy.nan,'Pune']
})

i.count(axis='columns')
```

```python
# to_excel() method in DataFrame
import pandas as pd
import numpy

i = pd.DataFrame({
    'Name':['John','Sam','Mac','Max'],
    'Age':[30,25,35,32],
    'Marks':[98,85,93,82],
    'Grade':['A','B','C','D'],
})
j = pd.ExcelWriter('dataofuser.xlsx')
i.to_excel(j)
j.save()
print('Excel File Generated.')
```
